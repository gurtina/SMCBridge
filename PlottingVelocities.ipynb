Plotting Data from Besla+2012 

Data is used in Zivick+2019 https://ui.adsabs.harvard.edu/abs/2019ApJ...874...78Z/abstract
Their figure 10

Here is example routine for loading and using the data files. 



Notes:
- COM Motion and Position of SMC is removed in all cases 

- In Model 2 the bridge is much shorter than observed, so you don't get as large of range of positions as in Model 1 

- there are no new stars formed in Model 1 in the bridge. 

- The data doesn't go straight to the SMC center - see example plots that will show the viewing perspective in the line of sight frame. In all cases the LMC is towards the bottom left of the images. 

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Old stars (>1 Gyr old)

Data = np.loadtxt('OldStars_Model2.txt')
pos = Data[:,0] # position along vector connecting the SMC center to that of the LMC center,
#projected on to plane perpendicular to our line of sight in kpc. 
#Values are negative, meaning closer to the LMC than the SMC center. 
# SMC COM motion removed from kinematics
pmx = Data[:,1]  # mean proper motion along direction from SMC center to the LMC center. Negative means towards the LMC.
pmy = Data[:,2]  # mean proper motion perpendicular to direction from SMC center to LMC center. Positive means north
vrad = Data[:,3] # mean radial velocity along line of sight, positive means away from us. 
sigmax = Data[:,4] # dispersion about mean pmx
sigmay = Data[:,5] # dispersion about mean pmy
sigmar = Data[:,6] # dispersion about mean vrad 
dist = Data[:,7] # mean heliocentric distance
sigmaD = Data[:,8] # dispersion about mean heliocentric distance
```

```python
# Old stars (>1 Gyr old)

Data1 = np.loadtxt('OldStars_Model1.txt')
pos1 = Data1[:,0] # position along vector connecting the SMC center to that of the LMC center,
#projected on to plane perpendicular to our line of sight in kpc. 
#Values are negative, meaning closer to the LMC than the SMC center. 
# SMC COM motion removed from kinematics
pmx1 = Data1[:,1]  # mean proper motion along direction from SMC center to the LMC center. Negative means towards the LMC.
pmy1 = Data1[:,2]  # mean proper motion perpendicular to direction from SMC center to LMC center. Positive means north
vrad1 = Data1[:,3] # mean radial velocity along line of sight, positive means away from us. 
sigmax1 = Data1[:,4] # dispersion about mean pmx
sigmay1 = Data1[:,5] # dispersion about mean pmy
sigmar1 = Data1[:,6] # dispersion about mean vrad 
dist1 = Data1[:,7] # mean heliocentric distance
sigmaD1 = Data1[:,8] # dispersion about mean heliocentric distance
```

```python

# New stars (< 1 Gyr old)

DataN = np.loadtxt('NewStars_Model2.txt')
posN = DataN[:,0] # position along vector connecting the SMC center to that of the LMC center,
#projected on to plane perpendicular to our line of sight in kpc. 
#Values are negative, meaning closer to the LMC than the SMC center. 
# SMC COM motion removed from kinematics
pmxN = DataN[:,1]  # mean proper motion along direction from SMC center to the LMC center. Negative means towards the LMC.
pmyN = DataN[:,2]  # mean proper motion perpendicular to direction from SMC center to LMC center. Positive means north
vradN = DataN[:,3] # mean radial velocity along line of sight, positive means away from us. 
sigmaxN = DataN[:,4]# dispersion about mean pmx
sigmayN = DataN[:,5]# dispersion about mean pmy
sigmarN = DataN[:,6]# dispersion about mean vrad
distN = DataN[:,7] # mean heliocentric distance
sigmaDN = DataN[:,8] # dispersion about mean heliocentric distance
```

```python
# Vlos vs. angular location along bridge

fig = plt.figure(figsize=(10,10))

ax = plt.subplot(111)

ax.scatter(pos, vrad, label='Old Stars Model 2')
ax.scatter(posN, vradN, label ='Young Stars Model 2')
ax.scatter(pos1, vrad1, label='Old Stars Model 1')

plt.xlabel('Distance in degrees from SMC (neg = towards LMC)')
plt.ylabel('Radial Velocity (km/s) relative to SMC COM Vrad')

plt.legend(loc='upper right')
```

```python
# Vlos vs. heliocentric distance

fig = plt.figure(figsize=(10,10))

ax = plt.subplot(111)

ax.scatter(dist, vrad, label='Old Stars Model 2')
ax.scatter(distN, vradN, label ='Young Stars Model 2')
ax.scatter(dist1, vrad1, label='Old Stars Model 1')

plt.xlabel('Heliocentric Distance (kpc) ')
plt.ylabel('Radial Velocity (km/s) relative to SMC COM Vrad')

plt.legend(loc='upper right')
```
